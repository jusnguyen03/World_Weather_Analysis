```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
from citipy import citipy
import numpy as np

# Import the API key.
from config import g_key

# Import the requests library.
import requests

from datetime import datetime
import time
from scipy.stats import linregress
```

```python
# Store the CSV you saved created in part one into a DataFrame.
city_data_df = pd.read_csv("weather_data/WeatherPy_challenge.csv")
city_data_df.head()
```

```python
# Create a column with Yes or No for rain
city_data_df['rain(y/n)'] = np.where(city_data_df['Rain inches(last 3hrs)']>0, 'yes', 'no')
city_data_df
```

```python
# Create a column with Yes or No for snow
city_data_df['snow(y/n)'] = np.where(city_data_df['Snow inches(last 3hrs)']>0, 'yes', 'no')
city_data_df
```

```python
#Get the data types
city_data_df.dtypes
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Ask the customer to add a minimum and maximum temperature value.Filter the DataFrame for minimum and maximum temperature preferences, and if the rain or snow accumulation is 0 inches or not using conditional statements.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
rain_q = str(input("Do you want it to be raining? (yes/no)"))
snow_q = str(input("Do you want it to be snowing? (yes/no)"))
```

```python
# Filter the dataset to find the cities that fit the criteria.
preferred_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & \
                                       (city_data_df["Max Temp"] >= min_temp) & \
                                       (city_data_df["rain(y/n)"] == rain_q) & \
                                       (city_data_df["snow(y/n)"] == snow_q)]
preferred_cities_df
```

```python
preferred_cities_df.count()
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
Hotelvacation_df = preferred_cities_df[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng"]].copy()
Hotelvacation_df["Hotel Name"] = ""
Hotelvacation_df
```

```python
# Set the parameters to search for a hotel in Paris.
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key,
    "location": "48.8566, 2.3522"}
# Use base URL to search for hotels in Paris.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
# Make request and get the JSON data from the search.
hotels = requests.get(base_url, params=params).json()
```

```python
 # Iterate through the DataFrame.
for index, row in Hotelvacation_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to the params dictionary as values to the location key.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        Hotelvacation_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
```
